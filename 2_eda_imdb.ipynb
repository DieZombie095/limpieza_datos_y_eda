```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("./data/imdb.csv")

print("HEAD DEFAULT")
print(df.head())

print("TAIL DEFAULT")
print(df.tail())

print("HEAD 10")
print(df.head(10))

print("TAIL 10")
print(df.tail(10))

print("COLUMNAS")
print(df.columns)

print("PRIMEROS RANK")
print(df["Rank"].head())

print("METASCORE CON CORCHETES")
print(df["Metascore"].head())

print("METASCORE CON PUNTO")
print(df.Metascore.head())

df.columns = df.columns.str.replace(" ", "_")

print("COLUMNAS RENOMBRADAS")
print(df.columns)

print(df.Metascore.head())
print(df.Rating.head())

print("INFO DATAFRAME")
print(df.info())

print("NA POR COLUMNA")
print(df.isna().sum())

print("NA EN METASCORE")
print(df["Metascore"].isna().sum())

print("NA EN TODAS LAS COLUMNAS")
for col in df.columns:
    print(col, df[col].isna().sum())

df = df.dropna()

print("INFO DESPUES DROPNA")
print(df.info())

print("ESTADISTICOS")
print(df.describe())

plt.hist(df["Metascore"], bins=10)
plt.title("Histograma Metascore")
plt.xlabel("Metascore")
plt.ylabel("Frecuencia")
plt.show()

plt.hist(df["Rating"], bins=10)
plt.title("Histograma Rating")
plt.xlabel("Rating")
plt.ylabel("Frecuencia")
plt.show()

print("DESCRIBE NUEVO")
print(df.describe())

print("PROMEDIO RATING NUMPY")
print(np.mean(df["Rating"]))

print("PROMEDIO RATING PANDAS")
print(df["Rating"].mean())

print("VALORES UNICOS RATING ORDENADOS")
print(np.sort(df["Rating"].unique()))

ratings = [7.5, 8.0, 8.5]
print("PELICULAS CON RATING SELECCIONADO")
print(df[df["Rating"].isin(ratings)])

rating_counts = df["Rating"].value_counts()

ratings_df = pd.DataFrame({
    "Rating": rating_counts.index,
    "Frecuencia": rating_counts.values
})

print("DATAFRAME FRECUENCIAS")
print(ratings_df)

ratings_df = ratings_df.sort_values(by="Rating")
print("DATAFRAME ORDENADO")
print(ratings_df)

plt.bar(ratings_df["Rating"], ratings_df["Frecuencia"])
plt.xlabel("Rating")
plt.ylabel("Frecuencia")
plt.title("Frecuencia de Ratings")
plt.show()

corr = df.corr(numeric_only=True)
print("MATRIZ CORRELACION")
print(corr)

plt.matshow(corr)
plt.colorbar()
plt.title("Matriz de Correlacion")
plt.show()
```
